### <center> ÁREA TÉCNICA 

### Unidad de competencia.
<p style="text-align: justify;"> El estudiante en un ambiente cooperativo investiga conocimientos teóricos básicos para intervenir en el contexto del dibujo electrónico, así como habilidades de investigación documental y de campo que le permita la comprensión de lo que se trata respecto al dibujo de una manera responsable y con trabajo de equipo. </p>


### Articulación de los ejes.
<p style="text-align: justify;"> Los estudiantes deben comprender las teorías del dibujo electrónico, desarrollando las competencias teórico- prácticas en la aplicación de problemas de diseño con apertura, disposición, compromiso, honestidad y respeto.</p>

### Descripción.
<p style="text-align: justify;"> Esta experiencia se localiza en el área de Iniciación de la disciplina (3 hrs. Taller, 3 créditos) el objetivo es brindar conocimientos a los estudiantes sobre conceptos básicos del dibujo electrónico, así como las nomenclaturas del dibujo a nivel de ingeniería y su visión integral en el que intervienen tres elementos de igual importancia entre sí que son: Tecnología, economía y sociedad. Para su desarrollo se proponen las estrategias metodológicas de simulación, reglas de diseño y elaboración de las tarjetas de circuito impreso. Por lo tanto, el desempeño de la unidad de competencia se evidencia mediante reportes de prácticas del software aplicado en el dibujo electrónico, así como exámenes, diseño e implementación de un circuito electrónico, y entrega de proyecto final.</p>

### Justificación.
<p style="text-align: justify;"> En la actualidad es de vital importancia para el ingeniero en electrónica y comunicaciones comprender lo significativo que es el dibujo electrónico en función del equilibrio que debe existir entre la tecnología, economía y sociedad. Los estudiantes deben aprender que los conceptos de este conocimiento, son la base fundamental de las experiencias educativas consecutivas. </p>

### Saber teorico.
- Representación gráfica de dibujo técnico.
- Simbología de componentes eléctricos y electrónicos.
- Normas y nomenclaturas de dibujo y de diseño electrónico.
- Herramientas de software para el dibujo electrónico
- Elaboración de tarjetas de circuitos impresos.
- Técnicas de transferencia de pistas a tablilla.
- Técnicas de montaje y soldadura.
- Diseño de chasis
- Proyecto final

### Saber heuristico.
- Interpretación de la información.
- Resolución de ejercicios.
- Uso de las TIC¿s.
- Uso de software para la elaboración de tablillas de circuito impreso.

### Saber axiologico.
- Responsabilidad entregando las evidencias de desempeño en tiempo y forma.
- Colaboración en el trabajo en equipos
- Respeto a sus compañeros y profesor.
- Profesionalismo en la presentación de los resultados.

### Estrategia metodológica de aprensizaje.
- Organización de grupos para trabajo en equipo
- Tareas para estudio independiente en clase y extractase.
- Aprendizaje basado en problemas de pistas, redes y componentes.

### Evidencia de desempeño.
- Exámenes.
- Realización de prácticas en el laboratorio y simulaciones a través de un software.
- Trabajo final.
